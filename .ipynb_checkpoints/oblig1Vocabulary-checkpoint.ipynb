```python
import numpy as np
import pandas as pd
import re
import regex
import nltk
import matplotlib.pyplot as plt
#nltk.download('stopwords')
from nltk.corpus import stopwords

data = pd.read_csv('train.csv', delimiter = ',')

pd.options.display.max_colwidth = 100

display(data.head(4))
```

```python
print(data.shape)
```

```python
filter = data['text'].str.contains('', regex = True, na = False)

data = data[filter]
print(data.shape)
```

```python
data['total']=data['title']+' '+data['author']+data['text']
```

```python
data=data.fillna('')
```

```python
print("before non-latin removal:", data.shape)
```

```python
filter = []
def latin_filter(field):
    for index, row in data.iterrows():
        reg = regex.match('^\p{Latin}', row[field])
        if reg == None:
            filter.append(True)
        else:
            filter.append(False)


latin_filter('total')
```

```python
filter = np.array(filter)

filter
```

```python
data = data[filter]
print("after non-latin removal:", data.shape)
```

```python
pd.options.display.max_colwidth = 100
display(data.head())
```

```python
data['text'].replace(regex=True, inplace=True,to_replace='\n', value='')
txts = data['text'].tolist()
```

```python
stopWords = set(stopwords.words('english'))
unique_words = []
def extract_words(text):
    global unique_words
    words = text.split(" ")
    
    for w in words:
        if unique_words.count(w.lower()) == 0 and not w == '' and w not in stopWords:
            unique_words.append(w.lower())
            
    
for text in txts:
    extract_words(text)
```
